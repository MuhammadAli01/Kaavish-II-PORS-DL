```
# Move directories
import os
import shutil

srcpath = '/content/drive/My Drive/Deep Fashion Retrieval/base/img_all'
srcfiles = os.listdir(srcpath)

destpath = '/content/drive/My Drive/Deep Fashion Retrieval/base/img'

categories = ['Anorak', 'Blazer', 'Bomber', 'Button-Down', 'Cardigan', 
              'Flannel', 'Henley', 'Hoodie', 'Jacket', 'Jersey', 'Parka', 
              'Peacoat', 'Sweater', 'Tank', 'Tee', 'Top', 'Turtleneck', 
              'Chinos','Jeans', 'Joggers', 'Shorts', 'Sweatpants', 
              'Sweatshorts', 'Trunks', 'Coat', 'Robe']

for cat_name in categories:
  to_move_lst = [dir_name for dir_name in srcfiles if dir_name.endswith(cat_name)]
  for to_move in to_move_lst:
    shutil.move(os.path.join(srcpath, to_move), destpath)
  print(f'Category {cat_name} moved: {to_move_lst}')
```
